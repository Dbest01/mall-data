```python
# importing seaborn
import seaborn as sns

# importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import IPython
from IPython.display import Image
```

## Relational plot API Interface

Relplot: Designed for visualizing different statistical relationships.

```python
# Loading Example Data
#tips = sns.load_dataset("tips")

df = pd.read_csv('Datasets/Blood Donation/transfusion.csv')
df.head(2)
```

```python
sns.relplot(
    data=df,
    x="Recency (months)", y = 'Frequency (times)', col= "whether he/she donated blood in March 2007",
    hue = 'Monetary (c.c. blood)', style="Time (months)"
)
```

```python
df1 = pd.read_csv('Datasets/Churn Analysis/WA_Fn-UseC_-Telco-Customer-Churn.csv')
sns.relplot(
    data=df1
)
plt.savefig("mall-analysis/images/relplot")
plt.show()
```

```python
import seaborn as sns
diamond = sns.load_dataset("diamonds")
diamond.head()
```

```python
sns.scatterplot(
    data= diamond,
    x='carat', y='price',
    hue='clarity'
)
#to save image as pdf
plt.savefig("mall-analysis/artifacts/caratvprice.pdf", format="pdf")
plt.show()
```

```python
#lineplot
sns.lineplot(
    data= diamond,
    x='carat', y='price',
    hue='clarity'
)
plt.savefig("mall-analysis/images/caratvprice", format="pdf")
plt.show()
```

# Distributional plots(Displot)

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
#displot
sns.displot(
    data= penguins,
    x='island', col='species',

)
```

```python
#Histplot
sns.histplot(
    diamond, x ='price', hue = 'cut'
    )
plt.savefig("mall-analysis/images/cutvprice")
plt.show()
```

```python
# KDE plot
sns.displot(
    diamond, x ='price', kind = 'kde'
    )
```

# categorical plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# catplot
sns.catplot(
    data=tips,
    x='day',
    y='total_bill'
)
plt.savefig("mall-analysis/images/dayvtotalbill")
plt.show()
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill'
    kind='swarm'
)
```
